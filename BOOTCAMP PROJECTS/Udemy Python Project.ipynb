```python
udemy =pd.concat([df1, df2,df3,df4], axis=0, ignore_index=True)
udemy.head()
```

```python
udemy2 =udemy.copy()
```

```python
udemy2=udemy2.drop(['published_date','Day','Free_Paid', 'Beginner_Course_status','Revenue','Day2'], axis =1)
```

```python
#convert from string to datetime
udemy2['published_timestamp']=pd.to_datetime(udemy2['published_timestamp'],utc=True)
```

```python
udemy2.info()
```

```python
 # Splitting the date part from DataFrame
udemy2['Published_Date'] = udemy2['published_timestamp'].dt.date
udemy2.head(1)
```

```python
#Check the weekday(Name)
udemy2['Day'] = udemy2['published_timestamp'].dt.day_name().str[:3]
udemy2.head(2)
```

```python
#Check if the course is free or paid
udemy2['Free_Paid'] = ['Free' if k==0  else 'Paid' for k in udemy2.price]
udemy2.head(1)
```
